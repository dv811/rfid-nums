```python
df['rfid']=df['rfid'].astype(str)
df['first']=df['rfid'].str.rpartition('.')[0]
df['hex1']=df['first'].astype(int).apply(hex).astype(str)
df['hex1']=list(map(lambda x: x[2:], df['hex1']))
df['last']=df['rfid'].str.rpartition('.')[2]
df['hex2']=df['last'].astype(int).apply(hex).astype(str)
df['hex2']=list(map(lambda x: x[2:], df['hex2']))
```

```python
for i in range(0,len(df['last'])):
    if df['last'][i][0]=='0':
        df['hex2'][i]='0'+df['hex2'][i]
```

```python
df['hex']=df['hex1']+df['hex2']
df['dec']=list(map(lambda x : int(x,16),df['hex']))
```
